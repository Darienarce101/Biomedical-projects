$\textbf{Importamos librerias}$

```python
import cv2 #
import matplotlib.pyplot as plt
import numpy as np
import importlib
from PIL import Image
```

$\textbf{Parametros}$

```python
plt.rcParams['figure.dpi'] = 200 #Aumentamos dimensiones de la imagen a un 200%
```

$\textbf{Visualización de imagenes}$

```python
img = 'Tesla.jpg'
Tesla_color = cv2.imread(img, cv2.IMREAD_COLOR)[:, :, ::-1]
plt.imshow(Tesla_color)
plt.title('Nikola Tesla a color')
plt.axis('off')
```

```python
# Caracteristicas
img_max=np.amax(Tesla_color)
img_min=np.amin(Tesla_color)
img_dim=np.shape(Tesla_color)

print(f'Las dimensiones de la imagen son {img_dim}')
print(f'La intensidad del pixel más oscuro es {img_min}')
print(f'La intensidad del pixel más claro es {img_max}')
print(f'La clase de la imagen es {type(Tesla_color).__name__}')
print(f'La clase de cada pixel es {Tesla_color.dtype}')
```

$\textbf{Histograma}$

```python
plt.hist(Tesla_color.ravel(), 256)
plt.title('Histograma')
plt.xlabel('Intensidad del pixel')
plt.ylabel('Cantidad de pixeles');
```

```python
color = ('r','g','b')
for i, c in enumerate(color):
    hist = cv2.calcHist([Tesla_color], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0,256])
    plt.title('Histogramas por canales')
    plt.xlabel('Intensidad del pixel')
    plt.ylabel('Cantidad de pixeles');

plt.show()
```

```python
plt.subplot(1, 3, 1)
plt.hist(Tesla_color[:,:,0].ravel(), 256)
plt.title('Histograma C. Rojo')
plt.xlabel('Intensidad del pixel')
plt.ylabel('Cantidad de pixeles');

plt.subplot(1, 3, 2)
plt.hist(Tesla_color[:,:,1].ravel(), 256)
plt.title('Histograma C. Verde')
plt.xlabel('Intensidad del pixel')
plt.ylabel('Cantidad de pixeles');

plt.subplot(1, 3, 3)
plt.hist(Tesla_color[:,:,2].ravel(), 256)
plt.title('Histograma C. Azul')
plt.xlabel('Intensidad del pixel')
plt.ylabel('Cantidad de pixeles');

```
